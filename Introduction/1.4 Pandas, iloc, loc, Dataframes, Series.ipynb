```python
# Импорт сторонних библиотек
import pandas as pd
import numpy as np
```

```python
# Считываем учебный файл с данными студентов
students = pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
```

```python
# Первые и последние 5 строк датасета
students
```

```python
# Первые 5 строк датасета
students.head()
```

```python
# Последние 5 строк датасета
students.tail()
```

```python
# Описательная статистика по всему датасету, исключая NaN значения
students.describe(include="all")
```

```python
# Информация о данных, включая dtype столбцов, количество ненулевых значений и использование памяти
students.info()
```

```python
# Типы данных столбцов
students.dtypes
```

```python
# Количество строк и столбцов
students.shape
```

```python
# Количество элементов в датасете
students.size
```

```python
# Получение элементов по индексам строк и столбцов
students.iloc[0:5,2:3]
```

```python
# Создание датасета из существующего
st_names = students.iloc[[1,3,5],[0,2]]
# Задание названий строк
st_names.index = [['A', 'B', 'C']]
```

```python
st_names
```

```python
# Получение элементов по названиям строк и столбцов
st_names.loc[['A', 'B'], ["gender"]]
```

```python
# Серия из 0 столбца
students.iloc[:,0]
```

```python
# Вывод типа данных
type(students.iloc[:,0])
```

```python
# Вывод столбца по названию как серию
students['gender']
```

```python
# Вывод столбца по названию как датафрейм
students[['gender']]
```

```python
# Создание серии
pd.Series([1, 2, 3])
```

```python
# Создание двух серий
seria_1 = pd.Series([1, 2, 3], index=['A', 'B', 'C'])
seria_2 = pd.Series([2, 4, 6], index=['A', 'B', 'C'])
```

```python
# Преобразование серий в датафрейм
pd.DataFrame({'col_name_1':seria_1, 'col_name_2':seria_2})
```
