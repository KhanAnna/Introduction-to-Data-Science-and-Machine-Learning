```python
# Скачаем датасэт с данными о героях из игры dota 2
dota=pd.read_csv('https://stepik.org/media/attachments/course/4852/dota_hero_stats.csv')
```

```python
# Посмотрим на распределение их возможных ролей в игре (колонка roles)
# Построим гистограмму, отражающую скольким героям сколько ролей приписывается (по мнению Valve, конечно) и 
# напишем какое число ролей у большинства героев
dota
```

```python
dota.describe(include='all')
```

```python
dota['count_roles']=dota.roles.str.split().str.len()
dota
```

```python
# Группировка количество ролей - количество героев
dota.groupby('count_roles').aggregate({'id':'count'}).rename(columns={'id':'count'}).sort_values('count', ascending=False)
```

```python
# Гистограмма
dota['count_roles'].hist()
```

```python
# Одной строкой
dota.roles.apply(lambda k: len(eval(k))).hist()
```

```python
# Скачаем датасэт со значениями параметров ирисов
iris=pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv')
```

```python
iris
```

```python
# Построим их распределения
sns.kdeplot(data=iris, log_scale=True, warn_singular=False)
```

```python
for column in iris:
    sns.displot(iris, x=column, kde=True)
```

```python
'''
Рассмотрим длину лепестков (petal length) подробнее и воспользуемся для этого violin плотом. 
Нарисуем распределение длины лепестков ирисов из предыдущего датасэта с помощью violin плота

Ссылки для изучения:
violinplot: https://seaborn.pydata.org/generated/seaborn.violinplot.html
'''
sns.violinplot(x=iris["petal length"])
```

```python
# Какая из пар навскидку имеет наибольшую корреляцию?
# Первый способ поиска: sns.pairplot()
import seaborn as sns
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
# Ответ: пара petal width и petal length имеет наибольшую корреляцию
```

```python
# Второй способ: iris.corr()
iris.corr()
```

```python
# Третий способ: sns.heatmap()
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True, cmap= 'coolwarm')
```

```python
# Вариации sns.pairplot()
sns.set(style="ticks", color_codes=True)
sns.pairplot(iris,  kind = "reg")
```
