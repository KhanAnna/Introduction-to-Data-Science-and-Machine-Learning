```python
# Импорт сторонних библиотек
import pandas as pd
import numpy as np
```

```python
# Считаем данные о героях из Dota 2
heroes = pd.read_csv('https://stepik.org/media/attachments/course/4852/dota_hero_stats.csv')
```

```python
# Первые и последние 5 строк датасета
heroes
```

```python
# Описательная статистика по всему датасету, исключая NaN значения
heroes.describe(include='all')
```

```python
# Пересчитаем число ног у героев игры Dota2
# Сгруппируем героев из датасэта по числу их ног (колонка legs)
heroes.groupby('legs').count()
# Число героев с 0 ног - 11, с 2-мя ногами - 95, с 4-мя ногами - 7, с 6-ю ногами - 3, с 8-ю ногами - 1
```

```python
# Продолжим исследование героев Dota2
# Сгруппируем по колонкам attack_type и primary_attr и выберем самый распространённый набор характеристик
heroes.filter(items=['attack_type','primary_attr']).mode()
# Ответ: самый распространённый набор характеристик - Ranged int
```

```python
# Считаем данные о заработках Лупы и Пупы за разные задачи
income = pd.read_csv('https://stepik.org/media/attachments/course/4852/accountancy.csv')
```

```python
# Первые 5 строк датасета
income.head()
```

```python
# Посмотрим, у кого из них больше средний заработок в различных категориях (колонка Type) и заполним таблицу, 
# указывая исполнителя с большим заработком в каждой из категорий
income.groupby(['Type', 'Executor'], as_index=False).aggregate({'Salary': 'mean'}).rename(columns={'Salary': 'Mean_salary'})
```

```python
income.drop('Unnamed: 0', axis=1) \
    .groupby(['Type', 'Executor'], as_index=False) \
    .mean() \
    .sort_values(['Type', 'Salary'], ascending=[True, False]) \
    .groupby('Type') \
    .head(1)
```

```python
# Продолжим исследование героев Dota2. Сгруппируем по колонкам attack_type и primary_attr и 
# выберем самый распространённый набор характеристик.
```

```python
# Первые 5 строк датасета
heroes.head()
```

```python
heroes.groupby(['attack_type', 'primary_attr'], as_index=False).count()
```

```python
# Группировка по столбцам attack_type и primary_attr + агрегация count по id
res = heroes.groupby(['attack_type', 'primary_attr'], as_index=False).aggregate({'id':'count'})
```

```python
# Переименование нового столбца и сортировка строк по убыванию count
res.rename(columns = {'id':'count'}).sort_values('count', ascending=False)
```

```python
# Скачаем данные о водорослях (вид каждой водоросли, её род (группа, объединяющая близкие виды), 
# группа (ещё одно объединение водорослей в крупные фракции) и концентрация анализируемых веществ.)
cnt = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
cnt.head()
```

```python
cnt.describe()
```

```python
''' 
Найдём среднюю концентрацию каждого из веществ в каждом из родов (колонка genus) 
Для этого проведем группировку датафрэйма, сохранённого в переменной concentrations, 
и применим метод, сохранив результат в переменной mean_concentrations
'''
mean_concentrations = cnt.groupby(['genus']).mean()
mean_concentrations
```

```python
'''
Пользуясь предыдущими данными, укажем через пробел (без запятых) чему равны минимальная, средняя и максимальная 
концентрации аланина (alanin) среди видов рода Fucus. Округлим до 2-ого знака, десятичным разделителем является точка.
'''
cnt.query("genus == 'Fucus'")
```

```python
cnt.query("genus == 'Fucus'").describe()
```

```python
# min, mean, max по столбцу alanin для водорослей рода genus == 'Fucus'
res = cnt.query("genus == 'Fucus'").alanin.describe().loc[['min', 'mean', 'max']]
res
```

```python
# Округление до двух знаков после запятой
np.around(res.values, decimals=2)
```

```python
# Сгруппируем данные по переменной group и выясним следующие параметры:
# group_count - число видов бурых, зелёных и красных водорослей
# citrate_var - дисперсия концентрации цитрата в бурых, зелёных и красных водорослях
# sucrose_range - размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях

# Первый способ - краткий
cnt.groupby('group') \
   .agg(
      group_count = ('group', 'count'),
      citrate_var = ('citrate', 'var'),
      sucrose_range = ('sucrose', lambda x: x.max() - x.min())) \
   .round(2)
```

```python
# Второй способ - по частям
# Число видов бурых, зелёных и красных водорослей
cnt.groupby('group').count()
```

```python
# Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях
def dif(lab):
    return np.max(lab)-np.min(lab)
cnt.groupby('group').agg({'sucrose':dif})
```

```python
# std - стандартное отклонение
cnt.groupby('group').citrate.describe(include='all')
```

```python
# Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях (квадрат стандартного отклонения)
(cnt.groupby('group').citrate.std().values)**2
```
