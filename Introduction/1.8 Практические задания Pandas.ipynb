### Задание 6 - замена наблюдений в данных

В dataframe с именем my_stat сохранены данные с 4 колонками: session_value, group, time, n_users.  

В переменной session_value замените все пропущенные значения на нули.

В переменной n_users замените все отрицательные значения на медианное значение переменной n_users (без учета отрицательных значений).

### Решение

````python
# Импорт библиотек
import pandas as pd    

# Скачиваем данные
my_stat = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat.csv')

# Переименовываем столбцы
my_stat.columns = ['session_value','group','time','n_users']

# В переменной session_value заменяем все пропущенные значения на нули
my_stat.fillna(value={"session_value": 0}, inplace=True)

# В переменной n_users заменяем все отрицательные значения на медианное значение переменной n_users
my_stat.loc[my_stat['n_users'] < 0, 'n_users']=my_stat[my_stat.n_users >= 0.0].n_users.median()

# Вывод
my_stat
````

### Задание 7 - группировка и агрегация данных
В этой задаче для данных my_stat рассчитайте среднее значение переменной session_value для каждой группы (переменная group), в получившемся dataframe  переменная group не должна превратиться в индекс. Также переименуйте колонку со средним значением session_value в mean_session_value.

Получившийся результат сохраните в dataframe с именем mean_session_value_data.

### Решение

````python
# Импорт библиотек
import pandas as pd    

# Скачиваем данные
my_stat = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat.csv')

# Переименовываем столбцы
my_stat.columns = ['session_value','group','time','n_users']

mean_session_value_data = my_stat \
    .groupby("group",as_index=False) \
    .session_value \
    .agg({'mean_session_value':'mean'})

# Вывод
mean_session_value_data
````
